```python

import cv2

import numpy as np
from tensorflow.keras.models import load_model
from mongoengine import connect, Document, StringField
```

```python
emotion_dict = {0: 'Angry', 1: 'Disgust', 2: 'Fear', 3: 'Happy', 4: 'Neutral', 5: 'Sad', 6: 'Surprise'}
```

```python
emotion_model = load_model('modelFinalPfa.h5')
```

```python
emotion_model.predict()
```

```python
face_detector = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
def process_video(input_video_path):
    cap = cv2.VideoCapture(input_video_path)
    if not cap.isOpened():
        print("nonono")
        return None

    emotion_counts = {emotion: 0 for emotion in emotion_dict.values()}
    total_frames = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        total_frames += 1
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_detector.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=10)

        for (x, y, w, h) in faces:
            roi_gray_frame = gray_frame[y:y+h, x:x+w]
            gray_input = cv2.resize(roi_gray_frame, (48, 48))
            rgb_input = cv2.cvtColor(gray_input, cv2.COLOR_GRAY2RGB)  # Convert to 3 channels (RGB)
            rgb_input = np.expand_dims(rgb_input, axis=0)
            rgb_input = rgb_input / 255.0

            try:
                print("ani wkeft")
                emotion_prediction = emotion_model.predict(rgb_input)
                print("waywa taadit")
                maxindex = int(np.argmax(emotion_prediction))
                emotion_counts[emotion_dict[maxindex]] += 1
            except Exception as e:
                print(f"Error during prediction: {e}")

    cap.release()
    if total_frames == 0:
        return None
    emotion_percentages = {emotion: (count / total_frames) * 100 for emotion, count in emotion_counts.items()}
    return emotion_percentages
```
